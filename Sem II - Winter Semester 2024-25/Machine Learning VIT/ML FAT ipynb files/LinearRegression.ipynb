#Linear Regression Model#
#Step 1: Importing necessary Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


data =({
    "Study_hours": [2,4,6,8,10],
    "Test_score": [50,55,60,70,85],
})
df=pd.DataFrame(data)
print(df)

x=df["Study_hours"]
y=df["Test_score"]
print(x)
print(y)

#Step3: Model Training

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
x_train=pd.DataFrame(x_train)
y_train=pd.DataFrame(y_train)
model=LinearRegression()
model.fit(x_train,y_train)

#Step 4: Get Regression Line:-  y=mx+c

m=model.coef_[0].item()
c=model.intercept_.item()
print(f"Regression Line: Y = {m:.2f}X+{c:.2f}")
```

#calculate predicted value

```python
x_reshaped = np.array(x).reshape(-1,1)
x_reshaped_df = pd.DataFrame(x_reshaped, columns=['Study_hours'])
y_predicted = model.predict(x_reshaped_df)
df['Predicted Score']=y_predicted

#Step5: Calculate Error and accuracy

print(df)

mae=mean_absolute_error(y,y_predicted)
mse=mean_squared_error(y,y_predicted)
r2=r2_score(y,y_predicted)
print(mae)
print(mse)
print(r2)
```

#Step 7: Plotting on Graph

```python
plt.scatter(x,y,color="blue",label="Actual Score")
plt.plot(x,y_predicted,color="red",label="Predicted Score")
plt.xlabel('Study Hours')
plt.ylabel('Test Score')
plt.title('Actual Score vs Predicted Score Regression Model')
plt.legend()
plt.show()
```
