```python
import numpy as np

# Define the dataset (X: independent variable, y: dependent variable)
# Example: X = [1, 2, 3, 4, 5], y = [1, 2, 2.8, 4.2, 5.1]
X = np.array([1, 2, 3, 4, 5])  # Independent variable (feature)
y = np.array([1, 2, 2.8, 4.2, 5.1])  # Dependent variable (target)

# Step 1: Calculate the mean of X and y
mean_X = np.mean(X)
mean_y = np.mean(y)

# Step 2: Calculate the coefficients (slope 'm' and intercept 'b')
# Slope (m) calculation using the formula: m = Σ((X_i - mean_X) * (y_i - mean_y)) / Σ((X_i - mean_X)^2)
numerator = np.sum((X - mean_X) * (y - mean_y))
denominator = np.sum((X - mean_X) ** 2)
slope = numerator / denominator

# Intercept (b) calculation using the formula: b = mean_y - m * mean_X
intercept = mean_y - (slope * mean_X)

# Step 3: Make predictions (y_pred) using the regression equation: y = m * X + b
y_pred = slope * X + intercept

# Print the calculated slope (m) and intercept (b)
print(f"Slope (m): {slope}")
print(f"Intercept (b): {intercept}")

# Print the predictions
print("Predictions: ", y_pred)

# Step 4: Calculate the Mean Squared Error (MSE) to assess the model
mse = np.mean((y - y_pred) ** 2)
print(f"Mean Squared Error (MSE): {mse}")
```

```python
import numpy as np

# Define Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Define Bipolar Sigmoid function
def bipolar_sigmoid(z):
    return 2 / (1 + np.exp(-z)) - 1

# Define ReLU function
def relu(z):
    return np.maximum(0, z)

# Define the dataset (X: features, y: labels)
X = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])  # Features
y = np.array([0, 1, 0, 1])  # Labels (binary classification)

# Add a column of ones to X for the intercept term (bias)
X_b = np.c_[np.ones(X.shape[0]), X]  # Add bias term

# Initialize weights (theta) to zeros
theta = np.zeros(X_b.shape[1])

# Learning rate and number of iterations for gradient descent
learning_rate = 0.01
iterations = 1000

# Define a function to compute the cost using the sigmoid function
def compute_cost(X, y, theta, activation_func):
    m = len(y)
    predictions = activation_func(np.dot(X, theta))  # Apply the activation function
    cost = -1/m * (y.dot(np.log(predictions)) + (1 - y).dot(np.log(1 - predictions)))
    return cost

# Define a function to perform gradient descent
def gradient_descent(X, y, theta, learning_rate, iterations, activation_func):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        predictions = activation_func(np.dot(X, theta))  # Apply the activation function
        gradients = (1/m) * X.T.dot(predictions - y)  # Compute the gradients
        theta -= learning_rate * gradients  # Update the weights

        # Calculate the cost for every iteration to monitor convergence
        cost = compute_cost(X, y, theta, activation_func)
        cost_history.append(cost)

    return theta, cost_history

# Train the logistic regression model using different activation functions

# Using Sigmoid function
print("Training with Sigmoid function...")
theta_sigmoid, cost_history_sigmoid = gradient_descent(X_b, y, theta, learning_rate, iterations, sigmoid)
print("Theta (weights) after training with Sigmoid:", theta_sigmoid)

# Using Bipolar Sigmoid function
print("\nTraining with Bipolar Sigmoid function...")
theta_bipolar_sigmoid, cost_history_bipolar_sigmoid = gradient_descent(X_b, y, theta, learning_rate, iterations, bipolar_sigmoid)
print("Theta (weights) after training with Bipolar Sigmoid:", theta_bipolar_sigmoid)

# Using ReLU function
print("\nTraining with ReLU function...")
theta_relu, cost_history_relu = gradient_descent(X_b, y, theta, learning_rate, iterations, relu)
print("Theta (weights) after training with ReLU:", theta_relu)

# Plotting cost history to visualize convergence (optional)
import matplotlib.pyplot as plt

plt.plot(cost_history_sigmoid, label="Sigmoid")
plt.plot(cost_history_bipolar_sigmoid, label="Bipolar Sigmoid")
plt.plot(cost_history_relu, label="ReLU")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.legend()
plt.title("Cost Function Convergence")
plt.show()
```

```python
import numpy as np

# Gaussian Probability Density Function
def gaussian_pdf(x, mean, std):
    return (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std) ** 2)

# Naive Bayes Classifier (Gaussian Naive Bayes)
class NaiveBayes:
    def __init__(self):
        self.class_probs = {}
        self.mean = {}
        self.std = {}

    # Fit the model
    def fit(self, X, y):
        self.classes = np.unique(y)
        self.class_probs = {c: np.sum(y == c) / len(y) for c in self.classes}  # P(C)
        self.mean = {}
        self.std = {}
        
        # Calculate mean and std for each feature for each class
        for c in self.classes:
            X_c = X[y == c]  # Extract the samples for class c
            self.mean[c] = np.mean(X_c, axis=0)
            self.std[c] = np.std(X_c, axis=0)
    
    # Predict the class for a given sample X
    def predict(self, X):
        predictions = [self._predict(x) for x in X]
        return np.array(predictions)

    # Predict for a single sample
    def _predict(self, x):
        # Calculate the posterior probability for each class
        posteriors = []
        for c in self.classes:
            prior = np.log(self.class_probs[c])  # log(P(C))
            likelihood = np.sum(np.log(gaussian_pdf(x, self.mean[c], self.std[c])))  # log(P(X|C))
            posterior = prior + likelihood  # Posterior = log(P(C)) + log(P(X|C))
            posteriors.append(posterior)
        
        # Return the class with the highest posterior
        return self.classes[np.argmax(posteriors)]

# Example usage
if __name__ == "__main__":
    # Example dataset
    X = np.array([[1.0, 2.0], [2.0, 3.0], [3.0, 4.0], [4.0, 5.0], [5.0, 6.0]])
    y = np.array([0, 0, 1, 1, 1])  # Class labels (0 or 1)

    # Initialize and train the Naive Bayes classifier
    nb = NaiveBayes()
    nb.fit(X, y)

    # Make predictions on new data
    X_new = np.array([[2.5, 3.5], [4.5, 5.5]])
    predictions = nb.predict(X_new)

    print("Predictions:", predictions)
```

```python
import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Artificial Neural Network class without backpropagation
class ANN:
    def __init__(self, input_size, hidden_size, output_size):
        # Initialize weights and biases
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        
        # Initialize weights and biases with random values
        # Input layer to hidden layer weights
        self.W1 = np.random.randn(self.input_size, self.hidden_size) * 0.1  
        self.b1 = np.zeros((1, self.hidden_size))  # Bias for hidden layer
        
        # Hidden layer to output layer weights
        self.W2 = np.random.randn(self.hidden_size, self.output_size) * 0.1  
        self.b2 = np.zeros((1, self.output_size))  # Bias for output layer
    
    def forward(self, X):
        # Forward pass through the network (no backpropagation)
        # Input to hidden layer
        self.Z1 = np.dot(X, self.W1) + self.b1
        self.A1 = sigmoid(self.Z1)  # Activation of hidden layer
        
        # Hidden layer to output layer
        self.Z2 = np.dot(self.A1, self.W2) + self.b2
        self.A2 = sigmoid(self.Z2)  # Activation of output layer
        
        return self.A2
    
# Example usage
if __name__ == "__main__":
    # Simple dataset: XOR problem
    # Input data (XOR inputs)
    X = np.array([[0, 0],
                  [0, 1],
                  [1, 0],
                  [1, 1]])
    
    # Output data (XOR outputs)
    y = np.array([[0], [1], [1], [0]])

    # Define the neural network
    input_size = 2  # Number of input features (XOR has 2 inputs)
    hidden_size = 4  # Number of neurons in the hidden layer
    output_size = 1  # XOR output has 1 value (0 or 1)

    # Initialize the ANN
    ann = ANN(input_size, hidden_size, output_size)
    
    # Make predictions (no training, as there's no backpropagation)
    predictions = ann.forward(X)

    print("Predictions (No Training):")
    print(predictions)
```

```python
# LINEAR + MULTI

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, accuracy_score

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/ML_Datasets/svm_diabetes_data.csv')

# Preprocessing: Replace 0.5 with 0 in the target column
df['diabetes'] = df['diabetes'].replace(0.5, 0)

# Prepare features and target

# # Define features and target variable
# X = df.drop(columns=['diabetes'])  # Features
# y = df['diabetes']  # Target variable


X = df[['HbA1c_level', 'blood_glucose_level']].values  # Convert to numpy array
y = df['diabetes'].values  # Convert to numpy array

# Split data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#Manual Linear Regression (Normal Equation)**
# Add bias term (column of ones)
X_train_b = np.c_[np.ones((len(X_train), 1)), X_train]
X_test_b = np.c_[np.ones((len(X_test), 1)), X_test]

# Compute theta using Normal Equation
theta_best = np.linalg.inv(X_train_b.T.dot(X_train_b)).dot(X_train_b.T).dot(y_train)

# Extracting intercept and coefficients
intercept_manual = theta_best[0]
coefficients_manual = theta_best[1:]

# Predictions
y_pred_manual = X_test_b.dot(theta_best)

# Convert regression output to binary classification (threshold 0.5)
y_pred_manual_class = np.round(y_pred_manual)

# Accuracy score
manual_accuracy = accuracy_score(y_test, y_pred_manual_class) * 100

print("\n📌 **Manual Linear Regression Results:**")
print(f"Intercept: {intercept_manual}")
print(f"Coefficients: {coefficients_manual}")
print(f"Manual Linear Regression Accuracy: {manual_accuracy:.2f}%")

### 2️⃣ **Linear Regression Using `sklearn`**
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Predictions
y_pred_sklearn = lin_reg.predict(X_test)

# Convert regression output to binary classification (threshold 0.5)
y_pred_sklearn_class = np.round(y_pred_sklearn)

# Accuracy score
sklearn_accuracy = accuracy_score(y_test, y_pred_sklearn_class) * 100

print("\n📌 **Scikit-Learn Linear Regression Results:**")
print(f"Intercept: {lin_reg.intercept_}")
print(f"Coefficients: {lin_reg.coef_}")
print(f"Scikit-Learn Linear Regression Accuracy: {sklearn_accuracy:.2f}%")

# Error Deviation
# mae= mean_absolute_error(y_test, y_pred_class)
# mse = mean_squared_error(y_test, y_pred_class)
# r2= r2_score(y_test, y_pred_class)
# print(mae)
# print(mse)
# print(r2)
```
